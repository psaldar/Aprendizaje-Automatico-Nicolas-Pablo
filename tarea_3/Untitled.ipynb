```python
import sqlite3
import pandas as pd

city = 'bogota'
query = f"""SELECT Datetaken, Longitude, Latitude, City
            FROM dataMeshFilt
            WHERE City = '{city}'"""
            
conn = sqlite3.connect("C:/Users/pasal/Google Drive/EAFIT/Ingenieria Matematica/09. Noveno Semestre/turism_and_qol/data/output/city_flickr_data_mesh.sqlite3")
data = pd.read_sql_query(query, conn)
```

```python
data.plot.scatter(x = 'Longitude', y='Latitude')
```

```python
data['year'] = data['Datetaken'].dt.year
```
