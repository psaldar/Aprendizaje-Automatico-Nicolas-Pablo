```python
import numpy as np
import pandas as pd
from random import randint
import matplotlib.pyplot as plt
```

```python
info_umbrales = pd.read_csv('data/sim_umbrales.csv', sep =',')

info_test = pd.read_csv('data/sim_umbrales_test.csv', sep =',')
```

```python
porcentajes = [0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5]
data = info_umbrales.drop(columns = '0').values
```

```python
n,m = data.shape
```

```python
B = 1000
lower_bound = []
upper_bound = []
for i in range(m):
    res_var = []
    info_aux = data[:,i]
    for b in range(B):
        idx = randint(0,m-1)
        data_aux_rm = np.delete(info_aux, idx)
        res_var.append(data_aux_rm.mean())
    lower_bound.append(np.percentile(res_var,2.5))
    upper_bound.append(np.percentile(res_var,97.5))
```

```python
summary = pd.DataFrame()
summary['lower_bound'] = lower_bound
summary['upper_bound'] = upper_bound
summary['test'] = info_test
summary['Porcentajes'] = porcentajes
```

```python
ax = summary.plot(x='Porcentajes')
```
